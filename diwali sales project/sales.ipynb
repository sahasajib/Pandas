```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
pd.isnull(df)
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
#drop unrelated /blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
df.shape
```

```python
#initialize list of lists
data_test = [['Sajib',11], ['Gopi', 15],['Keshab',], ['Anik',16]]
#create the pandas dataframe using list
df_test = pd.DataFrame(data_test, columns=['Name', 'Age'])

df_test
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.info()
```

```python
df.rename(columns= {'Marital_Status' : 'Shaadi'})
```

```python
df.describe()
```

```python
sns.countplot(x = 'Gender', data = df)
```

```python
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender', y= 'Amount', data = Sales_gen)
```

From above graphs we can see that most of the buyers are female and even the purchasing power of female are greater than men

```python
df.columns
```

```python
ax = sns.countplot(x = 'Age Group', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group', y= 'Amount', data = Sales_age)
```

```python
sns.countplot(x = 'Age Group', data = df, hue = 'Gender')
```

From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

```python
Sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x = 'State', y= 'Orders', data = Sales_state)
```

```python
Sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x = 'State', y= 'Amount', data = Sales_state)
```

From above graphs we can see that most of the orders are from uttar pradesh, maharashtra and karnataka respectively but total sales/amount is from up karanata and then maharastra

```python
ax = sns.countplot(x = 'Marital_Status', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_status = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x = 'Marital_Status', y= 'Amount', data = Sales_status, hue= 'Gender')
```

From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

```python
ax = sns.countplot(x = 'Occupation', data = df, hue= 'Gender')
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_state = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x = 'Occupation', y= 'Amount', data = Sales_state)
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x = 'Product_Category', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_state = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x = 'Product_Category', y= 'Amount', data = Sales_state)
```

From above graphs we can see that most of the sold product are from Food, Clothing & Apparel, Electronics gadgets category

```python
Sales_state = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x = 'Product_ID', y= 'Orders', data = Sales_state)
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
